```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
vocabulary_size = 10000
num_tags = 100
num_departements = 4

title = tf.keras.Input(shape = (vocabulary_size,), name = 'title')
text_body = tf.keras.Input(shape = (vocabulary_size,), name = 'text_body')
tags = tf.keras.Input(shape = (num_tags,), name = 'tags')


features = tf.keras.layers.Concatenate()([title, text_body, tags])
features = tf.keras.layers.Dense(64, activation = 'relu') (features)


priority = tf.keras.layers.Dense(1, activation = 'sigmoid', name = 'priority') (features)

departement = tf.keras.layers.Dense(num_departements, activation = 'softmax', name = 'departement') (features)

model = tf.keras.Model( inputs = [title, text_body, tags], outputs = [priority, departement])

tf.keras.utils.plot_model(model)
```

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist

(train_data, train_label), (test_data, test_label) = mnist.load_data()
```

```python
train_data = train_data.reshape((60000, 28, 28, 1)).astype('float32') / 255
test_data = test_data.reshape((10000, 28, 28, 1)).astype('float32') / 255
```

```python
train_data.shape
```

```python
def get_model() :
  inputs = tf.keras.Input(shape = (28, 28, 1))
  dense = tf.keras.layers.Conv2D(filters = 32, kernel_size= 3, activation = 'relu', padding = 'same') (inputs)
  dense = tf.keras.layers.MaxPooling2D(pool_size = 2) (dense)
  dense = tf.keras.layers.MaxPooling2D(pool_size = 2) (dense)
  dense = tf.keras.layers.Conv2D(filters = 64, kernel_size= 3, activation = 'relu', padding = 'same') (dense)
  dense = tf.keras.layers.MaxPooling2D(pool_size = 2) (dense)
  dense = tf.keras.layers.Conv2D(filters = 64, kernel_size= 3, activation = 'relu', padding = 'same') (dense)
  dense = tf.keras.layers.MaxPooling2D(pool_size = 2) (dense)
  dense = tf.keras.layers.Conv2D(filters = 128, kernel_size= 3, activation = 'relu', padding = 'same') (dense)
  dense  = tf.keras.layers.Flatten() (dense)
  output = tf.keras.layers.Dense(10, activation = 'softmax') (dense)

  model = tf.keras.Model(inputs = inputs, outputs = output)
  model.compile(
      loss = 'sparse_categorical_crossentropy',
      optimizer = 'adam',
      metrics = ['accuracy']
  )
  return model
```

```python
model = get_model()

call_back_list = [
    tf.keras.callbacks.ModelCheckpoint(filepath = 'model.h5', monitor = 'val_loss', save_best_only = True,),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor = 'val_loss',
        factor = 0.2,
        patience = 2,
        verbose = 1,
        min_lr = 1e-6,
    )
]
```

```python
history = model.fit(
    train_data,
    train_label,
    epochs = 10,
    callbacks = call_back_list,
    validation_split = 0.1,
    shuffle = True,
    batch_size = 100
)
```

```python
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])

plt.subplot(1, 2, 2)
plt.plot(history.history['val_accuracy'])
```

```python
loss, gain = model.evaluate(test_data, test_label)
```
